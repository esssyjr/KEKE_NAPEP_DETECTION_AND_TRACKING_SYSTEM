```python
import supervision as sv
from ultralytics import YOLO
import cv2
from collections import defaultdict, deque
import numpy as np
```

```python
video_path = "KEKE_NAPEP_TRACKING_TEST.MP4"
```

```python
model = YOLO("KEKE_NAPEP.pt") 
```

```python
video_info = sv.VideoInfo.from_video_path(video_path)
```

```python
byte_track = sv.ByteTrack(frame_rate = video_info.fps)
frame_generator =  sv.get_video_frames_generator(video_path)
bounding_box_annotator = sv.BoundingBoxAnnotator(thickness=3)
label_annotator = sv.LabelAnnotator()
```

```python
for frame in frame_generator:
    result = model(frame)[0]
    detections = sv.Detections.from_ultralytics(result)
    detections = byte_track.update_with_detections(detections=detections)


    annotated_frame = frame.copy()
    annotated_frame = bounding_box_annotator.annotate(
    scene=annotated_frame, detections=detections)
    annotated_frame = label_annotator.annotate(
    scene=annotated_frame, detections=detections)


    resized_frame = cv2.resize(annotated_frame, (1200,1000))
    cv2.imshow("annotated_frame", resized_frame)
    if cv2.waitKey(1) == ord("q"):
            break

cv2.destroyAllWindows()
```
